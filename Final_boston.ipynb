```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_digits
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from scipy import stats
import numpy as np
from sklearn.model_selection import train_test_split #trainsplit
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
data = pd.read_csv('train.csv')
```

```python
#We use corelation matrix to get a basic idea of which columsn to select
data.corr(method='pearson')
```

```python
data = data.drop(['indus','chas','nox','dis','rad','tax'],axis = 1)
```

```python
z = np.abs(stats.zscore(data))
#print(z)
#print(np.where(z > 3))
data = data[(z < 2).all(axis=1)]
```

```python
columns_target=['medv'] 
columns_train = ['lstat',
'ptratio',
'crim',
'age',
'black',
'zn',
'rm'
]
#X = data.iloc[:, :-1].values
X = data[columns_train]
Y = data[columns_target]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = .3, random_state = 42)
```

```python
regressor = LinearRegression()
regressor.fit(X_train,Y_train)
a = regressor.predict(X_test)
```

```python
rms = sqrt(mean_squared_error(Y_test,a))
rms
```
